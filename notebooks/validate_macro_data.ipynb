# Validate Macro Data

This notebook checks the downloaded FRED macro data against the classified ETF universe.

```python
from pathlib import Path

import numpy as np
import pandas as pd

BASE_DIR = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()

ETF_PATH = BASE_DIR / "data" / "raw" / "etf_universe_classified.csv"
MACRO_PATH = BASE_DIR / "data" / "raw" / "macro_indicators_fred.csv"

etf = pd.read_csv(ETF_PATH)
macro = pd.read_csv(MACRO_PATH)
macro["date"] = pd.to_datetime(macro["date"], errors="coerce")

COUNTRY_EXCLUDED = {
    "Asia Pacific ex-Japan",
    "Developed ex-US",
    "Emerging Markets",
    "Europe",
    "Global",
    "Global ex-US",
    "Latin America",
    "No information",
    "Other",
}

country_universe = sorted(
    value
    for value in etf["primary_region"].dropna().astype(str).unique()
    if value not in COUNTRY_EXCLUDED
)
```

```python
summary = {
    "macro_rows": len(macro),
    "macro_columns": len(macro.columns),
    "macro_country_labels": macro["country_or_region"].nunique(),
    "macro_indicators": macro["indicator_name"].nunique(),
    "macro_series_ids": macro["series_id"].nunique(),
    "date_min": macro["date"].min(),
    "date_max": macro["date"].max(),
    "etf_country_labels_expected": len(country_universe),
}
pd.Series(summary)
```

```python
indicator_counts = (
    macro.groupby("indicator_name")
    .agg(rows=("series_id", "size"), countries=("country_or_region", "nunique"), series=("series_id", "nunique"))
    .sort_index()
)
indicator_counts
```

```python
coverage = (
    macro.groupby(["country_or_region", "indicator_name"])["series_id"]
    .nunique()
    .unstack(fill_value=0)
    .reindex(country_universe)
)
coverage
```

```python
missing_pairs = (
    coverage.stack()
    .rename("series_count")
    .reset_index()
    .query("series_count == 0")
    .rename(columns={"level_0": "country_or_region", "level_1": "indicator_name"})
)
missing_pairs.head(30)
```

```python
country_summary = (
    macro.groupby("country_or_region")
    .agg(
        indicators=("indicator_name", "nunique"),
        series=("series_id", "nunique"),
        first_date=("date", "min"),
        last_date=("date", "max"),
        rows=("date", "size"),
    )
    .sort_values(["indicators", "series", "rows", "country_or_region"], ascending=[True, True, True, True])
)
country_summary.head(20)
```

```python
series_dates = macro.sort_values(["series_id", "date"]).copy()
series_dates["diff_days"] = series_dates.groupby("series_id")["date"].diff().dt.days

freq_check = (
    series_dates.groupby(["country_or_region", "indicator_name", "series_id"])
    .agg(
        observations=("date", "size"),
        median_gap_days=("diff_days", "median"),
        min_date=("date", "min"),
        max_date=("date", "max"),
    )
    .reset_index()
)

def classify_frequency(days):
    if pd.isna(days):
        return "Single point"
    if days <= 35:
        return "Monthly-ish"
    if days <= 100:
        return "Quarterly-ish"
    return "Irregular / sparse"

freq_check["frequency_flag"] = freq_check["median_gap_days"].apply(classify_frequency)
freq_check.sort_values(["frequency_flag", "country_or_region", "indicator_name"])
```

```python
freq_check["frequency_flag"].value_counts()
```

```python
freq_check.query("frequency_flag != 'Monthly-ish'").sort_values(["frequency_flag", "country_or_region", "indicator_name"]).head(30)
```
