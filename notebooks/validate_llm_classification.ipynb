# Validate LLM ETF Classification

This notebook checks the outputs produced by the sync LLM classification run.

```python
from pathlib import Path

import pandas as pd

BASE_DIR = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()

ORIGINAL_PATH = BASE_DIR / "data" / "raw" / "etf_universe_ishares.csv"
CLASSIFIED_PATH = BASE_DIR / "data" / "raw" / "etf_universe_classified.csv"
AUDIT_PATH = BASE_DIR / "data" / "raw" / "etf_classification_audit.csv"

original = pd.read_csv(ORIGINAL_PATH)
classified = pd.read_csv(CLASSIFIED_PATH)
audit = pd.read_csv(AUDIT_PATH)

TARGET_COLS = ["primary_region", "style_focus", "size_focus", "sector_focus"]
UPDATED_COLS = ["primary_region", "developed_or_emerging", "style_focus", "sector_focus"]
```

```python
summary = {
    "original_rows": len(original),
    "classified_rows": len(classified),
    "audit_rows": len(audit),
    "successful_rows": int((audit["status"] == "ok").sum()),
    "error_rows": int((audit["status"] == "error").sum()),
    "avg_confidence": round(audit["confidence"].dropna().mean(), 3),
    "min_confidence": round(audit["confidence"].dropna().min(), 3),
}
pd.Series(summary)
```

```python
before_after_other = pd.DataFrame(
    {
        "before_other": [int(original[col].fillna("").eq("Other").sum()) for col in TARGET_COLS],
        "after_other": [int(classified[col].fillna("").eq("Other").sum()) for col in TARGET_COLS],
    },
    index=TARGET_COLS,
)
before_after_other["reduction"] = before_after_other["before_other"] - before_after_other["after_other"]
before_after_other
```

```python
changes = audit.assign(
    primary_region_changed=lambda df: df["original_primary_region"] != df["updated_primary_region"],
    developed_or_emerging_changed=lambda df: df["original_developed_or_emerging"] != df["updated_developed_or_emerging"],
    style_focus_changed=lambda df: df["original_style_focus"] != df["updated_style_focus"],
    sector_focus_changed=lambda df: df["original_sector_focus"] != df["updated_sector_focus"],
)

pd.Series(
    {
        "primary_region_changed": int(changes["primary_region_changed"].sum()),
        "developed_or_emerging_changed": int(changes["developed_or_emerging_changed"].sum()),
        "style_focus_changed": int(changes["style_focus_changed"].sum()),
        "sector_focus_changed": int(changes["sector_focus_changed"].sum()),
    }
)
```

```python
review_cols = [
    "ticker",
    "original_primary_region",
    "updated_primary_region",
    "original_style_focus",
    "updated_style_focus",
    "original_sector_focus",
    "updated_sector_focus",
    "confidence",
    "reasoning_short",
]

audit.sort_values(["confidence", "ticker"], ascending=[True, True])[review_cols].head(15)
```

```python
sample_review = classified.loc[
    classified["ticker"].isin(audit.sort_values("confidence").head(10)["ticker"]),
    ["ticker", "etf_name", "description"] + UPDATED_COLS,
].sort_values("ticker")
sample_review
```
